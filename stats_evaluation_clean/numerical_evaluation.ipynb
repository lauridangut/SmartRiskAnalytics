### Descripción de features en dataset application_train

1. SK_ID_CURR: ID del préstamo en nuestra muestra

# Evaluación estadística de variables numéricas

```python
data_num_v1.shape
```

```python
# Eliminación de valores nulos para evaluación estadística
data_num_v1_sinnull = data_num_v1.dropna()
data_num_v1_sinnull.shape
```

```python
# Prueba t de student 
from scipy.stats import ttest_ind # type: ignore

resultados = []
for col in data_num_v1_sinnull.columns:
    group_0 = data_num_v1_sinnull.loc[data["TARGET"] == 0, col]
    group_1 = data_num_v1_sinnull.loc[data["TARGET"] == 1, col]
    mean_group_0 = group_0.mean()
    mean_group_1 = group_1.mean()
    stat, p_value = ttest_ind(group_0, group_1)
    resultados.append({"Feature": col,
                       "Estadístico t": stat,
                       "p-value": p_value,
                       "Media Grupo 0": mean_group_0,
                       "Media Grupo 1": mean_group_1})
df_student = pd.DataFrame(resultados)
df_student
```

```python
df_student["Significativa"] = np.where(df_student["p-value"] < 0.05, "Si", "No")
df_student
```

```python
# Filtrado de df_student por columnas no significativas
col_nosig = df_student[df_student["Significativa"] == "No"]
lista_col_nosig = col_nosig["Feature"].tolist()
lista_col_nosig
```

```python
len(lista_col_nosig)
```

```python
# Lista de columnas estadísticamente significativas
lista_col_sig = [col for col in data_num_v1_sinnull if col not in lista_col_nosig]
lista_col_sig
```

```python
len(lista_col_sig)
```

```python
# Armado de DataFrame con variables numéricas estadísticamente significativas para analizar colinealidad
df_sig = data_num_v1_sinnull[lista_col_sig]
df_sig.shape
```

### Imputación de valores nulos

```python
def imputacion_mediana(df, columnas):
    diccionario = {}
    for col in columnas:
        mediana = df[col].median()
        df[col] = df[col].fillna(mediana)
        porcentaje = df[col].isna().sum()*100/df.shape[0]
        diccionario[col] = porcentaje
    return diccionario
```

### Manejo de outliers

```python
# Función para eliminar outliers
def clean_outliers(df, df_name, cols):
    df_clean = df.copy()  # Hacemos una copia del DataFrame original
    for colname in cols: #Utilizaremos IQR ya que es mas robusto y no asume distribucion normal
        column_values = df_clean[colname].values
        Q1 = np.percentile(column_values, 25)
        Q3 = np.percentile(column_values, 75)
        IQR = Q3 - Q1
        LOWER = (Q1 - IQR * 1.5)
        UPPER = (Q3 + IQR * 1.5)
        # Extraemos los index que estan dentro de los parametros
        indices = (df_clean[colname].values >= LOWER) & (df_clean[colname].values <= UPPER)
        # Filtramos el df
        df_clean = df_clean[indices]
        shape_old = df.shape[0]
        shape_clean = df_clean.shape[0]
        print(f"El dataset {df_name} pasó de tener {shape_old} registros a tener {shape_clean} luego de la limpieza de outliers.")
    return df_clean
```

```python
cols_out = ["OBS_30_CNT_SOCIAL_CIRCLE", "DEF_30_CNT_SOCIAL_CIRCLE", "AMT_INCOME_TOTAL"]
data_num_v2 = clean_outliers(data_num_v2, "data_num_v2", cols_out)
```
